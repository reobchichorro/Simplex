```python
import math
import copy
import pandas as pd
import numpy as np
import os
```

```python
myfolder = "wrongFiles/doneFiles/"
myresfolder = "res/"
answersheetFolder = "../reference/Simplex/res/"
```

```python
cols = ["input", "myres", "myobj", "trueres", "trueobj"]
pddict = {}
for col in cols:
    pddict[col] = []

for filename in os.listdir(myfolder):
    file = open(myresfolder+filename+".txt", "r").readlines()
    pddict["input"].append(filename.removesuffix(".mps.txt"))
    found = False
    for line in file:
        if "nan" in line:
            pddict["myres"].append("nan")
            pddict["myobj"].append(math.nan)
            found = True
            break
        elif "Infeasible" in line:
            pddict["myres"].append("Inf")
            pddict["myobj"].append("")
            found = True
            break
        elif "Unbounded" in line:
            pddict["myres"].append("Unb")
            pddict["myobj"].append("")
            found = True
            break
        elif "Optimal" in line:
            pddict["myres"].append("Opt")
            for line2 in file:
                if "Solution cost: " in line2:
                    pddict["myobj"].append(float(line2.removeprefix("Solution cost: ")))
                    found = True
            break
    if not found:
        print("aqui " + filename)
    asfile = open(answersheetFolder+filename+".txt", "r").readlines()
    # print(filename)
    for line in asfile:
        # if "nan" in line:
        #     pddict["trueres"].append("nan")
        #     pddict["trueobj"].append(math.nan)
        #     break
        if "Infeasible" in line:
            pddict["trueres"].append("Inf")
            pddict["trueobj"].append("")
            break
        elif "Unbounded" in line:
            pddict["trueres"].append("Unb")
            pddict["trueobj"].append("")
            break
        elif "Optimal" in line:
            pddict["trueres"].append("Opt")
            pddict["trueobj"].append(float(line.split()[1]))
            break
```
